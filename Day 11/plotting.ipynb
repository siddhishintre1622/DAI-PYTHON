```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
months = np.array(['jan','feb','march','april','may'])
sales = np.array([50,200,250,400,500])
#sales = np.array(50,500, size=(5,5))

#prod = np.array(['A','B','C','D','E'])
```

```python
plt.figure(figsize=(8,6))
plt.plot(months,sales, marker= "*", alpha=1 )
plt.title("Sales Data")
plt.xlabel("Months")
plt.ylabel("sales")
plt.legend()
plt.grid(True)
plt.show()

```

```python
months = np.array(['Jan','Feb','Mar','Apr','May'])
sales = np.array([50,70,100,50,200])

plt.figure(figsize=(8,6))
plt.bar(months, sales, color='purple')
plt.title("Sales Data")
plt.xlabel('months')
plt.ylabel('sales')
plt.show()
```

```python
plt.pie(sales,labels=months,shadow = 'True', autopct='%2.2f%%')
plt.title("Pie Chart Sales")
plt.legend()
```
