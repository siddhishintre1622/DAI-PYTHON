```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['A','B','C','D']
sizes = [25,35,20,20]
colors = ['red','yellow','pink','blue']
explode = [0.1,0.1,0,0]
```

```python
#Line plot

fig,axs = plt.subplots(2,2, figsize=(12,12))



x = np.arange(1,11)
y1 = np.array([1,3,4,7,9,12,15,18,19,22])
y2 = np.array([2,5,7,10,12,13,16,18,21,23])
axs[0,0].plot(x,y1, label = "Dataset 1", color = "blue", marker="o", linestyle="--")
axs[0,0].plot(x,y2, label = "Dataset 2", color = "red",marker = "*", linestyle ="-")
axs[0,0].set_title("Line plot")
axs[0,0].set_xlabel('X- axis')
axs[0,0].set_ylabel('Y-axis')
axs[0,0].legend()
axs[0,0].grid(True)


#bar plot
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [4,7,1,8]

axs[0,1].bar(categories, values, color='purple')
axs[0,1].set_title('Bar Chart Example')
axs[0,1].set_xlabel('Categories')
axs[0,1].set_ylabel('Values')

#ScatterPlot

scatter = axs[1,0].scatter(categories,values,c= colors, alpha=0.5, cmap='viridis')
axs[1,0].set_title('Scatter Plot Example')
axs[1,0].set_xlabel('X Axis')
axs[1,0].set_ylabel('Y Axis')
fig.colorbar(scatter, ax=axs[1,0])

#piechart
months = np.array(['Jan','Feb','Mar','Apr','May'])
sales = np.array([50,70,100,50,200])
plt.pie(sales,labels=months,shadow = 'True', autopct='%2.2f%%')
plt.title("Pie Chart Sales")
plt.legend()
```
