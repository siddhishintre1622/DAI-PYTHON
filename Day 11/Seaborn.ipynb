```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('iris')
print(df.head())
```

```python
#scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal_length', y='petal_width', hue= 'species', data = df, palette='Set2' ,s =100)
plt.title('Scatter Plot: Petal Length vs. Petal Width')
plt.show()
```

```python
#box plot
plt.figure(figsize=(8,6))
sns.boxplot(x='species', y='sepal_length', data=df, palette='Pastel1')
plt.title('Box Plot: Petal Length by Species')
plt.show()
            
```

```python
#violin plot
plt.figure(figsize=(8,10))
sns.violinplot(x='species',y='petal_length', data=df,palette='Pastel1')
plt.title("Violin Plot")
plt.show()
```

```python
#pair plot

sns.pairplot(df, hue='species', palette='Set1')
plt.suptitle('Iris Dataset',y=1)
plt.show()
```
