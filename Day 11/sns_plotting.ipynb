```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
months = np.array(['jan','feb','march','april','may'])
products = np.array(['A','B','C','D','E'])
sales = np.random.randint(50,100,size=(5))

data = {
    'months':months,
    'products':products,
    'sales':sales
}
prod = pd.DataFrame(data)
prod

```

```python
plt.figure(figsize=(5,5))
sns.lineplot(x='months',y='sales', data=prod , marker='o', color='blue' )
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x= 'months', y= 'sales', color='purple' , data = prod)
plt.title('Bar Chart Example')

plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.violinplot(x='months', y='products', data=prod,palette='Pastel1')
plt.title("Violin chart")
plt.show()
```
