```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class StockData:

    # date = None
    # open = None
    # high = None
    # low = None
    # close = None
    # volume = None
    df = ""
    def __init__(self):
        pass
        
    def load_data(self,filename):
        StockData.df = pd.read_csv(filename,nrows= 5)
        print(StockData.df)
    

    def sort_data(self,col_name):
        StockData.df.sort_values(by = col_name,inplace=True)
        print(StockData.df)

    def visualize_data(self,x,y):
        plt.plot(x,y,marker = "*", linestyle= "--")
        plt.show()

        
```
