```python
import pandas as pd
```

```python
df = pd.read_csv('country_vaccinations.csv')
```

```python
print (df)
```

```python
print(df.head())
```

```python
df['vaccination_rate']= (df['total_vaccinations']-df['people_vaccinated'])/2000000*100
print(df)
```

```python
df['daily_vaccinations'] = df['daily_vaccinations'].fillna(df['daily_vaccinations'].mean())
print(df['daily_vaccinations'])
```

```python
print(df)
```

```python
df = df.drop('source_name',axis=1)
print(df)
```

```python
sorted = df.sort_index(ascending=False)
print(sorted)
```

```python
sort_by_country = df.sort_values(by='country')
print(sort_by_country)
```

```python
sort_by_total_vaccinations = df.sort_values(by='total_vaccinations')
print(sort_by_total_vaccinations)
```

```python
df = df.rename(columns={'iso_code':'country_code'})
print(df)
```

```python
df= df.rename(index={'0':'1'})
print(df)
```

```python
df.index = df.index+1
print (df)
```

```python
missing_data= df.isnull().sum()
print(missing_data)
```

```python
df_filled = df['people_fully_vaccinated'].fillna(0)
print(df_filled)
```

```python
print("original:",df)
df_dropped = df.dropna()
print(df_dropped)
```

```python
df['vaccines'] = df['vaccines'].replace({'Moderna, Pfizer/BioNTech':'mRNA vaccine'})
print(df)
```
