```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
day = np.arange(1,21)
d1 = pd.DataFrame({'Days':day})
d1

```

```python
temp = np.random.uniform(15,31,size=(20))
d2 = pd.DataFrame({'Temperature':temp})
d2
```

```python
humidity  = np.random.randint(40,90,size=(20))
d3 = pd.DataFrame({'humidity' : humidity})
d3
```

```python
data = {
    'Days':day,
    'Temperature':temp,
    'humidity' : humidity    
}
data = pd.DataFrame(data)

data.to_csv('weather.csv')
data = pd.read_csv('weather.csv')
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x='Temperature', y='Days', marker='o', linestyle='--', color='red' ,data=data)
plt.title('Lineplot')
plt.show()
```

```python
sns.scatterplot(x='Temperature',y='humidity',data=data)
plt.show()
```
